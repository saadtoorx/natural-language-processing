```python
from fastapi import FastAPI, Form
import requests

app = FastAPI()

@app.post("/analyze/")
def analyze_sentiment(text: str = Form(...)):
    prompt = (
        "What is the sentiment of this text? "
        "Respond with Positive, Negative, or Neutral:\n\n"
        f"{text}"
    )

    response = requests.post(
        "http://localhost:11434/api/generate",
        json={
            "model": "mistral",
            "prompt": prompt,
            "stream": False
        }
    )

    result = response.json()
    return {"sentiment": result["response"].strip()}
```

```python
import streamlit as st

st.title("Sentiment Analyzer (Mistral)")

text_input = st.text_area("Enter your sentence here:")

if st.button("Analyze"):
    response = requests.post(
        "http://localhost:8000/analyze/",
        data={"text": text_input}
    )
    sentiment = response.json().get("sentiment", "Error")
    st.subheader("Predicted Sentiment:")
    st.write(sentiment)
```
